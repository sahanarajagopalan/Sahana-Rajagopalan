Section 4: Data Science

 Do people who listen to music more than three hours a day have less anxiety than people who listen to music less than three hours a day? I will use a one sided two sample Z test to compare both samples to eachother. I can use the z-test because the sample size of both groups is greater than 30. I will start with defining my two groups. My first group will be those who listen to music for less than 3 hours. My second group will be those who listen to music for 3 or more hours. Lastly, I will run the one sided two sampe Z test to look at the correlation of listening to more music and anxiety. 

```python
anxiety= "Anxiety"
hours_listened= "Hours per day"
less_hours_listened= df[df[hours_listened] <3]
less_hours_listened
more_hours_listened= df[df[hours_listened] >=3]
more_hours_listened
from statsmodels.stats.weightstats import ztest
anxiety_results = ztest(less_hours_listened[anxiety].dropna(), more_hours_listened[anxiety].dropna(), alternative = "larger")
anxiety_results
```
